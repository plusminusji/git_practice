### 정수 num1과 num2가 주어질 때, num1과 num2의 합을 return하도록 soltuion 함수를 완성해주세요.
### [제한사항]
* -50,000 ≤ num1 ≤ 50,000
* -50,000 ≤ num2 ≤ 50,000

```python
def solution(num1, num2):
    return num1 + num2
```

```python
solution(2,3)
```

```python
solution(100,2)
```

### 다른 사람 풀이

```python
lambda num1, num2: num1 + num2
```
