### 정수 num1, num2가 매개변수로 주어질 때, num1를 num2로 나눈 나머지를 return 하도록 solution 함수를 완성해주세요.
### [제한사항]
* 0 < num1 ≤ 100
* 0 < num2 ≤ 100

```python
def solution(num1, num2):
    a, b = divmod(num1, num2)
    return b
```

```python
solution(3, 2)
```

```python
solution(10, 5)
```

### 다른 사람 풀이


```python
def solution(num1, num2):
    return int.__mod__(num1, num2)
```
