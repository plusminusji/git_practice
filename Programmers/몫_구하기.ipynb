### 정수 num1, num2가 매개변수로 주어질 때, num1을 num2로 나눈 몫을 return 하도록 solution 함수를 완성해주세요.

##### [제한사항]
* 0 < num1 ≤ 100
* 0 < num2 ≤ 100

```python
def get_quotient(num1, num2):
    return num1//num2
```

```python
get_quotient(10, 5)
```

```python
get_quotient(7, 2)
```

### 다른 사람 코드 리뷰하기

```python
divmod(num1, num2) # 출력: (quotient, remainder)
```
